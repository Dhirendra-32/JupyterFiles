```python
#how to join multiple ,we have three data frame df1,df2,df3
```

```python
df1= pd.DataFrame(np.array([['a',5,10],['b',4,61],['c',24,19]]),columns=['name','attri1','attri2'])
```

```python
df2= pd.DataFrame(np.array([['a',6,11],['b',5,62],['c',25,20]]),columns=['name','attri1','attri2'])
```

```python
df3 = pd.DataFrame(np.array([['a',7,12],['b',6,63],['c',26,21]]),columns=['name','attri1','attri2'])
```

```python
df1,df2,df3
```

```python
df1
```

```python
df2
```

```python
#joining these three data frame with innr join 
```

```python
pd.merge(df1,df2, on='name')#joining two frame df1 and df2 based on one colmnb
```

```python
pd.merge(pd.merge(df1,df2, on='name',how = 'right'),df3,on ='name',how='right')
```

```python
#datafram name >>syntac
```

```python
(flight_data_frame>>select(X.origin , X.dest,X.hour)).head()#just like sequal syntax
```

```python
(flight_data_frame>>
select(~X.origin ,~X.dest,~X.hour))#mean exclusion of origin dest and hour
```

```python
(flight_data_frame>>mask(X.month==1 ,X.day==1,X.origin =='JFK',X.hour>10))
```

```python
(flight_data_frame>>arrange(X.distance,X.hour,ascending= True)).head(50)
```

```python
(flight_data_frame>>mutate(newdistance=X.distance/100,carrier_origin = X.carrier+X.origin)).head(70)
```

```python
((flight_data_frame>>mask(X.year==2013))>>mutate(newdistance=X.distance/100,carrier_origin = X.carrier+X.origin)).head(70)
```

```python
(flight_data_frame>>group_by(X.origin)).head()
```

```python
(flight_data_frame>>group_by(X.origin))>>summarise(mean_dis=X.distance.mean())
```

```python
(flight_data_frame>>mask(X.hour>10)>>mutate(speed=X.distance/(X.air_time*60))>>group_by(X.origin)>>summarise(mean_speed=X.speed.mean())>>arrange(X.mean_speed,ascending=False))
```
